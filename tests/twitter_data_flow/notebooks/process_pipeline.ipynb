```python
import stdflow as sf
from stdflow import StepRunner, Pipeline

countries = sf.var("countries", ["spain", "italy"])

ppl = Pipeline()

for country in countries:
    ppl.add_step("process_sex.ipynb", variables={"country": country})
    ppl.add_step("process_tweets.ipynb", variables={"country": country})
ppl
```

```python
ppl.run(progress_bar=False)
```
